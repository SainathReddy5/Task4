###### import necessary libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

###### merge those two csv files (after getting as dataframes, get them as a single dataframe)

```python
df1 = pd.read_csv('college_1.csv')
df2 = pd.read_csv('college_2.csv')
df = pd.concat([df1,df2])        
```

###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


```python
EE_df1=df[['Name','CodeKata Score']]
EE_df1 = EE_df1[EE_df1['CodeKata Score'] > 15000]
EE_df1.set_index('Name')
EE_df1.to_csv('Exceeded expectations.csv')
```

###### if  10000<codekata score<15000   (Reached_expectations.csv)



```python
RE_df = df[['Name','CodeKata Score']]
RE_df = RE_df[(RE_df['CodeKata Score'] < 15000) & (RE_df['CodeKata Score'] > 10000)]
RE_df.set_index('Name')
RE_df.to_csv('Reached_expectations.csv')
```

###### if  7000<codekata score<10000   (Needs_Improvement.csv)


```python
NI_df = df[['Name','CodeKata Score']]
NI_df = NI_df[(NI_df['CodeKata Score'] < 10000) & (NI_df['CodeKata Score'] > 7000)]
NI_df.reset_index(drop = True)
NI_df.to_csv('Needs_Improvement.csv')
```

###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
undf = df[['Name','CodeKata Score']]
undf = undf[undf['CodeKata Score'] < 7000]
undf.to_csv('Unsatisfactory.csv')
```

###### Average of previous week geekions vs this week geekions (i.e Previous Geekions vs CodeKata Score)

```python
avg = {
    'avg_prev_geekcoins' : df['Previous Geekions'].mean(),
    'avg_codekata_score' : df['CodeKata Score'].mean()
    }
avg_df = pd.DataFrame(avg,index=[1])
avg_df = avg_df.T
sns.barplot(x=avg_df.index,y=avg_df[1])
```
